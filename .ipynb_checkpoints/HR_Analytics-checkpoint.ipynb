```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
df['Attrition'].value_counts()
```

```python
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition")
plt.show()
```

Most employees stayed with the company, while a smaller percentage left, indicating an imbalanced target variable.

```python
sns.countplot(x='Department', hue='Attrition', data=df)
plt.xticks(rotation=15)
plt.title("Attrition by Department")
plt.show()
```

Identify which department experiences the highest employee attrition.

```python
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title("Attrition by Gender")
plt.show()
```

```python
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title("OverTime vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['MonthlyIncome'], bins=30)
plt.title("Monthly Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Key Business Insights

After EDA, summarize findings such as:

Employees working overtime are more likely to leave.
Some departments have higher attrition than others.
Employees with lower monthly income tend to leave more often.
Newer employees may have a higher attrition rate.
